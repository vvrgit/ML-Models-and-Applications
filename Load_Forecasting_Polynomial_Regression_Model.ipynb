# **Import Libraries**

```python
import numpy as np # For vector operation
import pandas as pd # For Data manipulation
import matplotlib.pyplot as plt # For plotting
import seaborn as sns #For data visualization
```

# **Import Data**

```python
load_data=pd.read_csv("PROJECT_Final.csv")
load_data.head()
```

# **Load Data Stat Inofrmation**

```python
print(load_data.shape) # details about number of samples and features
load_data.describe()
```

# **Data Normalization**

```python
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
load_data=ms.fit_transform(load_data)
print(load_data)
```

# **Data Pre-Processing**

```python
load_data_process=np.zeros((len(load_data)-24,2))
print(load_data_process.shape)
```

```python
k=24
for i in range(2160):
    load_data_process[i,0],load_data_process[i,1]=load_data[k-24],load_data[k]
    k=k+1
print(load_data_process)
```

# **Data Split**

```python
dataset=pd.DataFrame(data=load_data_process[0:,0:])
```

```python
X=dataset.iloc[:,0:1].values
Y=dataset.iloc[:,1:].values
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
print(X_test[0:5])
```

# **Polynomial Features Extraction**

```python
from sklearn.preprocessing import PolynomialFeatures

polynomial_features= PolynomialFeatures(degree=10)
x_poly = polynomial_features.fit_transform(X_train)
```

# **Polynomial Regression Model Training**

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_poly, y_train)
```

# **Metrics Calculation**

```python
x_poly_test = polynomial_features.fit_transform(X_test)
y_poly_pred = model.predict(x_poly_test)
print(y_test[0])
print(y_poly_pred[0])
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_poly_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_poly_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_poly_pred)))
```

# **Model Deployment**

```python
x=float(input("enter previous day load"))
xnorm=(x-3378)/(8842-3378)
xnorm=np.array(xnorm)
xnorm=xnorm.reshape(-1,1)
x_poly = polynomial_features.fit_transform(xnorm)
today_load=model.predict(x_poly)
today_load=today_load*(8842-3378)+3378
print(today_load)
```
